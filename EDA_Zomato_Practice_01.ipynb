```python
# Drop the duplicates

zomato_data.drop_duplicates(inplace=True)
```

```python
# remove all the nan values

zomato_data.dropna(inplace=True)
```

```python
zomato_data.info()
```

```python
# renaming the columns
"""
approx_cost(for two people) - cost
listed_in(type) - type
listed_in(city) - city
"""


zomato_data.rename({'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'},axis=1,inplace=True)
```

```python
zomato_data.head()
```

```python
zomato_data.drop('menu_item',inplace=True,axis=1)
```

```python
zomato_data.head()
```

```python
# convert cost from object to float

zomato_data.cost.unique()
```

```python
zomato_data['cost'] = zomato_data['cost'].apply(lambda x : x.replace(',',''))
zomato_data['cost'] = zomato_data['cost'].astype(float)
```

```python
zomato_data.head()
```

```python
# convert the rating from decimal represntation to float representation

zomato_data.rate.unique()
```

```python
zomato_data = zomato_data[zomato_data.rate != "NEW"]
```

```python
zomato_data['rate'] = zomato_data['rate'].apply(lambda x : x.replace('/5',''))
```

```python
zomato_data['rate'].unique()
```

```python
# zomato_data

# online_order
# book_table

# convert Yes and No to True and False

```

```python
zomato_data.head()

zomato_data[['online_order','book_table']]=zomato_data[['online_order','book_table']].applymap(lambda x : True if x=='Yes' else False)
```
