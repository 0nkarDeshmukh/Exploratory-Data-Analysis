### Find most recent and earliest travelled date and mean distance travelled for each start city

```python
df=df.iloc[:-1,:]
df.tail()
```

```python
# Find most recent and earliest travelled date and mean distance travelled for each start city

res = df.groupby('START*').agg({'MILES*':['mean'],'START_DATE*':['min','max']})
res
```

```python
res = res.reset_index()
res
```

```python
res.columns = ['START_LOCATION','AVG_DIST','EARLIEST_TRIP','RECENT_TRIP']
```
